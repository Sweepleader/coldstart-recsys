测试VGGish的导入效果并测试对音频解析效果

```python
import torch

# 使用本地的hubconf.py文件加载模型
# 带有hubconf.py文件的上一级位置
vggish_dir = str('../VGGish/hub/harritaylor_torchvggish_master')

model = torch.hub.load(vggish_dir, 'vggish', source="local") # 本地加载时一定要加上source="local", 否则会报错, 因为源文件默认是从github下载的
model.eval()
```

```python
import torch
import torch.nn.functional as F

# 如果你已经执行完了github网站上的代码, 那么下面这条会直接从刚才的下载路径读取模型
model = torch.hub.load('harritaylor/torchvggish', 'vggish')
model.eval()

filename = './bus_chatter.wav'

with torch.no_grad():
    emb = model.forward(filename)          # [T, 128]
    vec = emb.mean(0)                      # [128]
    vec = vec.to(torch.float32) / 255.0
    vec = (vec - vec.mean()) / (vec.std() + 1e-6)
    vec = F.normalize(vec, dim=0)

print(emb.shape)
print(vec.shape)
print(vec)

"""
输出样式：

Using cache found in C:\###\###/.cache\torch\hub\harritaylor_torchvggish_master
torch.Size([19, 128])
torch.Size([128])
tensor([ 0.0445, -0.1316,  0.0236, -0.0711,  0.1687, -0.0649, -0.0006, -0.0798,
        -0.0287,  0.0871,  0.1229, -0.1106, -0.0007,  0.0622,  0.0290, -0.1013,
        -0.1258, -0.0116,  0.0638,  0.0931, -0.1545,  0.1690,  0.0119,  0.0240,
         0.0417, -0.1262,  0.0628,  0.0394, -0.0750, -0.0828,  0.0403, -0.0490,
        -0.0825,  0.0434, -0.0778, -0.0238, -0.0517, -0.0682, -0.0511, -0.0226,
         0.0200,  0.0368, -0.0426,  0.0962,  0.0451,  0.0781, -0.0812, -0.0514,
        -0.0639,  0.1120,  0.1330, -0.1071,  0.0488,  0.0308, -0.0730, -0.0534,
         0.0162, -0.0513,  0.0047,  0.0265,  0.1133, -0.0944, -0.0460,  0.1001,
         0.0715,  0.0570,  0.0121, -0.0202,  0.1330, -0.1418, -0.0946,  0.0872,
         0.1278,  0.1466, -0.0820,  0.0086, -0.0080, -0.0695, -0.0213,  0.1548,
        -0.0082,  0.0435,  0.0854, -0.0193,  0.0546, -0.0896,  0.0854, -0.0134,
         0.0326, -0.0885, -0.1448,  0.1547,  0.0270,  0.1071,  0.1176,  0.1012,
        -0.0234, -0.1474,  0.0705,  0.0486,  0.0779, -0.1748, -0.0838, -0.0818,
        -0.1628,  0.0741,  0.0059,  0.1685,  0.0973, -0.0247,  0.1545, -0.0904,
        -0.1718, -0.0651, -0.0465, -0.0263, -0.0256,  0.1028, -0.1561,  0.0802,
        -0.0117, -0.0863, -0.1695, -0.1127,  0.1092,  0.0227, -0.0663,  0.1755],
       device='cuda:0')
"""
```
